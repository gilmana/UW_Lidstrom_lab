```python
FM86 = pd.read_csv("FM86_csv.csv")
FM87_FM91 = pd.read_csv("FM91_rev1_csv.csv")
FM88 = pd.read_csv("FM88_v1_csv.csv")
FM90 = pd.read_csv("FM90_csv.csv")
```

```python
#plotting FM86
import matplotlib.patches as mpatches

figure_gas = plt.figure(figsize = (10,7))

# Twin the x-axis twice to make independent y-axes.
ax = figure_gas.add_subplot(111)
axes = [ax, ax.twinx(), ax.twinx(), ax.twinx()]

#adjust figure size as fraction of total figure (mesured from left and bottom)
#whitespace on right of figure that is 25% of figsize 
figure_gas.subplots_adjust(right=0.8, left = 0.20)

# Move the last y-axis spine over to the right by 20% of the width of the axes
axes[-1].spines['right'].set_position(('axes', 1.15))
#axes[-2].spines['right'].set_position(("axes", 1.15))

# To make the border of the right-most axis visible, we need to turn the frame
# on. This hides the other plots, however, so we need to turn its fill off.
axes[-1].set_frame_on(True)
axes[-1].patch.set_visible(False)


#plotting the OD
axes[0].scatter(FM86.Day_OD, FM86.CDW_g, s = 35, color="Black")
axes[0].set_ylabel('CDW (g)')
axes[0].tick_params(axis='y', colors="Black")

axes[0].set(xlim=[0, 10.5], ylim=[0, 0.65])

axes[0].set_xlabel('Time (day)')

#plotting the CH4
axes[1].scatter(FM86.Day_GC, FM86.CH4_mmol_hr, s=5, color="red")
#axes[1].set_ylabel('Methane Uptake (mmol per hour)', color = "Red")
#axes[1].tick_params(axis='y', colors="black")

axes[1].set(xlim=[0, 10.5], ylim=[0.6, 1.2])

#plotting the O2
axes[2].scatter(FM86.Day_GC, FM86.O2_mmol_hr, s=5, color="Blue")
axes[2].set_ylabel(r'Gas Uptake $(mmol\quad hr^{-1})$', color = "Black")
axes[2].tick_params(axis='y', colors="black")
axes[2].set(xlim=[0, 10.5], ylim=[0.6, 1.2])


#plotting the formate 
axes[3].scatter(FM86.Day_OD, FM86.Formate_umolar, s=35, color="green")
axes[3].set_ylabel(r'[Formate] ($\mu$M)', color = "green")
axes[3].tick_params(axis='y', colors="green")
axes[3].set(xlim=[0, 10.5], ylim=[0, 170])


#legend?
red_patch = mpatches.Patch(color='red', label='Methane uptake rate')
blue_patch = mpatches.Patch(color='blue', label='Oxygen uptake rate')

plt.legend(handles=[red_patch, blue_patch])
#plt.legend(handles=[blue_patch],loc = 2)

plt.title('FM86 5GB1 WT')
#figure_gas.legend(handles=[red_patch])
#ax2.legend(loc=0)

#save figure
figure_gas.savefig('FM86 5GB1-WT.png', bbox_inches='tight')
```

```python
#Solution for multiple y axis from stack exchange (matplotlib: overlay plots with different scales?) 
#answer by Joe Kington 

import numpy as np
np.random.seed(1977)

fig, ax = plt.subplots()

# Twin the x-axis twice to make independent y-axes.
axes = [ax, ax.twinx(), ax.twinx()]

# Make some space on the right side for the extra y-axis.
fig.subplots_adjust(right=0.75)

# Move the last y-axis spine over to the right by 20% of the width of the axes
axes[-1].spines['right'].set_position(('axes', 1.2))

# To make the border of the right-most axis visible, we need to turn the frame
# on. This hides the other plots, however, so we need to turn its fill off.
axes[-1].set_frame_on(True)
axes[-1].patch.set_visible(False)

# And finally we get to plot things...
colors = ('Green', 'Red', 'Blue')
for ax, color in zip(axes, colors):
    data = np.random.random(1) * np.random.random(10)
    ax.plot(data, marker='o', linestyle='none', color=color)
    ax.set_ylabel('%s Thing' % color, color=color)
    ax.tick_params(axis='y', colors=color)
axes[0].set_xlabel('X-axis')

plt.show()
```
